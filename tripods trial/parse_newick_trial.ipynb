```python
from ete3 import Tree

def split_newick_tree_from_file(file_path):
    with open(file_path, 'r') as file:
        newick_string = file.read().strip()  # Read the Newick tree string from the file

    return split_newick_tree(newick_string)


def split_newick_tree(newick_string):
    tree = Tree(newick_string, format=1)  # Parse the Newick tree string
    split_trees = []

    # Iterate over each clade and extract the subtree string
    for clade in tree:
        subtree = clade.write(format=1)
        split_trees.append(subtree)

    return split_trees


# Example usage
file_path = 'C:/Users/jyu1/Downloads/trial/C. elegans larva/size using original branch lengh/elegans2.txt'
split_trees = split_newick_tree_from_file(file_path)

# Print each subtree list
for subtree in split_trees:
    print(subtree)
```
