```python
Phylo.convert("elegans2.txt", "newick", "elegans2.xml", "phyloxml")
tree = Phylo.read("elegans2.xml", "phyloxml")
net1 = Phylo.to_networkx(tree)
data = list({"id": target.name, "name": target.name, "parent": src.name, "size": i } for i,(src,target) in enumerate(net1.edges()))
root = list(net1.edges())[0][0].name;
data = [{"id": root, "name": root }] + data
print(data)
```
